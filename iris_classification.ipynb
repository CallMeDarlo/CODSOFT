```python
#Removing Outliers
copy=df.copy()
def remove_outlier(data,features):
    output=data.copy()
    for i in features:
        q1=output[i].quantile(0.25)
        q3=output[i].quantile(0.75)
        iqr=q3-q1
        lower=q1-(1.5*iqr)
        upper=q3+(1.5*iqr)
        output.loc[output[i]<lower,i]=lower
        output.loc[output[i]>upper,i]=upper
    return output

features=['sepal_length','sepal_width','petal_length','petal_width']
df=remove_outlier(df,features)
```

```python
corr=df.copy()
corr=corr.drop('species',axis=1)
plt.figure(figsize=(12,8))
sns.heatmap(corr.corr(method='pearson'),annot=True)
plt.show()
```

**Model Training**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
x=df[['sepal_length','sepal_width','petal_length','petal_width']]
y=df[['species']]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=0.8)
```

```python
scaler=StandardScaler()
x_train_norm=scaler.fit_transform(x_train)
x_test_norm=scaler.fit_transform(x_test)
```

```python
model = LogisticRegression()
model.fit(x_train_norm,y_train)
```

**Model Evaluation**

```python
y_pred=model.predict(x_test_norm)
print("Accuracy of the model is : ",model.score(x_test_norm,y_test))
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title("Confusion Matrix")
plt.show()
```

```python
print("classification Report")
print(classification_report(y_test,y_pred))
```
